Extract text from one colored pdf

```python
import fitz  # PyMuPDF

def extract_highlighted_text(pdf_path):
    doc = fitz.open(pdf_path)
    highlighted_text = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        annotations = page.annots()
        if annotations:
            for annot in annotations:
                if annot.type[0] == 8:  # Highlight annotation
                    quadpoints = annot.vertices
                    quads = [quadpoints[i:i + 4] for i in range(0, len(quadpoints), 4)]
                    for quad in quads:
                        rect = fitz.Quad(quad).rect
                        words = page.get_text("words", clip=rect)
                        words.sort(key=lambda w: (w[1], w[0]))  # sort by y, then by x
                        highlighted_text.append(" ".join(w[4] for w in words))

    return highlighted_text

pdf_path = '/content/350351_1_highlighted.pdf'
highlights = extract_highlighted_text(pdf_path)
for text in highlights:
    print(text)
```

Extract text from multiple colored pdf

```python
import fitz  # PyMuPDF

def get_color_str(color):
    # Convert the color tuple to a string representation
    return f"rgb({color[0]*255:.0f},{color[1]*255:.0f},{color[2]*255:.0f})"

def extract_highlighted_text_by_color(pdf_path):
    doc = fitz.open(pdf_path)
    highlighted_text_by_color = {}

    for page_num in range(len(doc)):
        page = doc[page_num]
        annotations = page.annots()
        if annotations:
            for annot in annotations:
                if annot.type[0] == 8:  # Highlight annotation
                    color = annot.colors['stroke'] if 'stroke' in annot.colors else annot.colors['fill']
                    color_str = get_color_str(color)
                    if color_str not in highlighted_text_by_color:
                        highlighted_text_by_color[color_str] = []
                    quadpoints = annot.vertices
                    quads = [quadpoints[i:i + 4] for i in range(0, len(quadpoints), 4)]
                    for quad in quads:
                        rect = fitz.Quad(quad).rect
                        words = page.get_text("words", clip=rect)
                        words.sort(key=lambda w: (w[1], w[0]))  # sort by y, then by x
                        highlighted_text_by_color[color_str].append(" ".join(w[4] for w in words))

    return highlighted_text_by_color

pdf_path = '/content/350351_1_highlighted different.pdf'
highlights_by_color = extract_highlighted_text_by_color(pdf_path)
for color, texts in highlights_by_color.items():
    print(f"Highlights in color {color}:")
    for text in texts:
        print(f" - {text}")
```

Extract pdf page numbers that highlighetd in color rgb(143,222,249)

```python
import fitz  # PyMuPDF

def get_color_str(color):
    # Convert the color tuple to a string representation
    return f"rgb({color[0]*255:.0f},{color[1]*255:.0f},{color[2]*255:.0f})"

def extract_highlighted_page_numbers(pdf_path, target_color):
    doc = fitz.open(pdf_path)
    highlighted_page_numbers = []

    for page_num in range(len(doc)):
        page = doc[page_num]
        annotations = page.annots()
        if annotations:
            for annot in annotations:
                if annot.type[0] == 8:  # Highlight annotation
                    color = annot.colors['stroke'] if 'stroke' in annot.colors else annot.colors['fill']
                    color_str = get_color_str(color)
                    if color_str == target_color:
                        quadpoints = annot.vertices
                        quads = [quadpoints[i:i + 4] for i in range(0, len(quadpoints), 4)]
                        for quad in quads:
                            rect = fitz.Quad(quad).rect
                            words = page.get_text("words", clip=rect)
                            words.sort(key=lambda w: (w[1], w[0]))  # sort by y, then by x
                            highlighted_page_numbers.append("".join(w[4] for w in words))

    return highlighted_page_numbers

def extract_text_from_pages(pdf_path, page_numbers):
    doc = fitz.open(pdf_path)
    text_by_page = {}

    for page_number in page_numbers:
        try:
            page_index = int(page_number) - 1  # Convert to zero-based index
            if 0 <= page_index < len(doc):
                page_text = doc[page_index].get_text()
                text_by_page[page_number] = page_text
        except ValueError:
            # Handle cases where the page number is not an integer
            continue

    return text_by_page

pdf_path = '/content/350351_1_highlighted different.pdf'
target_color = 'rgb(143,222,249)'  # The color of the highlighted page numbers
highlighted_page_numbers = extract_highlighted_page_numbers(pdf_path, target_color)
text_by_page = extract_text_from_pages(pdf_path, highlighted_page_numbers)

for page_number, text in text_by_page.items():
    print(f"Text on page {page_number}:")
    print(text)
    print("\n")
```
